```python
# Redefine working dataframe to df
df = df1.copy()
```

```python
# Seaborn 

#define figure size and Seaborn style
sns.set(rc={"figure.figsize":(6, 4)}) #width=12, height=8
sns.set_style("whitegrid")

# Scatterplot
sns.scatterplot(data=df, x="age", y="fare", hue="sex", style="survived").set(
    xlabel="age", ylabel="fare", title="Seaborn Scatter Plot");
```

```python
# With Seaborn objects
from seaborn import axes_style
    
(
    so.Plot(
        df, x="age", y="fare", color="sex", marker="survived"
    )
    .add(so.Dot()).layout(size=(6, 4)).theme({**axes_style("whitegrid"), "grid.linestyle": ":"})
    .label(
        x="age", y="fare",title="Seaborn objects scatterplot"
    )
)
```

```python

# Using Seaborn barplot

sns.barplot(x=df['embark_town'],y=df['fare']).set(ylabel= "fare", title="Seaborn Bar Plot");

```

```python
# Seaborn objects barplot
(
    so.Plot(df, x="embark_town", y="fare", color="embark_town")
    .add(so.Bar(), so.Agg())
    .label(
        x="Embark town", y="fare",title="Seaborn objects plot"
    )
)
```

```python
# Using Seaborn barplot

sns.barplot(data=df, x='embark_town',y='fare', hue='survived', dodge=True).set(ylabel="Fare", 
                                                                                   title="Seaborn Dodged Bar Plot");
```

```python
# Seaborn objects barplot
(
    so.Plot(df, x="embark_town", y="fare", color="sex")
    .add(so.Bar(), so.Agg(), so.Dodge())
    .label(
        x="embark_town", y="Fare",title="Seaborn objects dodged scatterplot"
    )
)
```

```python
# Seaborn histogram

sns.histplot(df['age']).set(
    xlabel="age", title="Seaborn Histogram Plot");
```

```python
# Seaborn objects histogram

(
    so.Plot(df, x="fare")
    .add(so.Bars(), so.Hist()).label(
        x="Fare",title="Seaborn objects histogram"
    )
)
```

```python
# Define basic plot and add properties
p = so.Plot(df, "fare", color="embark_town")
p.add(so.Bars(), so.Hist()).label(
        x="fare",title="Seaborn objects histogram (non-stacked)"
    )
```

```python
p.add(so.Bars(), so.Hist(), so.Stack(), color="sex").label(
        x="Fare",title="Seaborn objects histogram (stacked)"
    )
```

```python
sns.set(rc={"figure.figsize":(4, 4)})
g = sns.FacetGrid(df, col="embark_town",  row="survived")
g.map_dataframe(sns.histplot, x="age");
```

```python
(
    so.Plot(df, x="fare")
    .facet(col="embark_town", row="sex")
    .add(so.Bars(), so.Hist()).label(
        x="fare"
    ).layout(size=(10, 6))
)
```

## Not yet implemented in seaborn.objects

```python
# Seaborn boxplots

sns.boxplot(data=df, x="survived", y="fare").set(
    ylabel="Fare", title="Seaborn Boxplot Plot");
```

```python
# Seaborn violin plot

sns.violinplot(data=df, x='embark_town', y='fare', hue='sex', height=12);
```

```python
# Seaborn violin plot

sns.violinplot(data=df, x='sex', y='fare', hue='survived',
               split=True, inner="quart", linewidth=1);
```

```python
# Seaborn Masked correlation heatmap 
#corr = df.corr()
# Select only numeric columns
corr = df.select_dtypes(include=np.number).corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
  f, ax = plt.subplots(figsize=(8, 8))
  ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True, annot=True, cmap='Blues');
```

```python
# Seaborn KDE plots

sns.FacetGrid(df, hue="sex", height=6,) \
   .map(sns.kdeplot, "age", fill=True) \
   .add_legend();
```

```python
# Seaborn pairplot using KDE plots
sns.pairplot(df, hue='sex', height=3);
```

```python
# Pairplot using histograms
sns.pairplot(df, hue="embark_town", height=3, diag_kind="hist");
```
